```python
# Check Missing Values
df["patient_nbr"].isnull().sum()
```

```python
# Check Duplicate Values
df["patient_nbr"].duplicated().sum()
```

```python
# Convert Data Type
df["patient_nbr"] = df["patient_nbr"].astype("int64")
```

## **3. race**

```python
# Check Data Type
df["race"].dtype
```

```python
# Check Unique Values
df["race"].unique()
```

```python
# Check Missing Values
df["race"].isnull().sum()
```

```python
# Replace '?' with NaN and Fill Missing Values with Mode
import numpy as np

df["race"] = df["race"].replace("?", np.nan)

mode_race = df["race"].mode()[0]

df["race"] = df["race"].fillna(mode_race)
```

```python
# Verify Missing Values
df["race"].isnull().sum()
```

```python
# Convert Categorical Values into Numerical Values
race_map = {
    "Caucasian": 1,
    "AfricanAmerican": 2,
    "Asian": 3,
    "Hispanic": 4,
    "Other": 5
}

df["race"] = df["race"].map(race_map)
```

```python
# Verify Encoding
df["race"].unique()
```

## **4. age**

```python
# Check Data Type
df["age"].dtype
```

```python
# Check Unique Values
df["age"].unique()
```

```python
# Check Missing Values
df["age"].isnull().sum()
```

```python
# Convert Age Groups into Numerical Values
age_map = {
    "[0-10)": 5,
    "[10-20)": 15,
    "[20-30)": 25,
    "[30-40)": 35,
    "[40-50)": 45,
    "[50-60)": 55,
    "[60-70)": 65,
    "[70-80)": 75,
    "[80-90)": 85,
    "[90-100)": 95
}

df["age"] = df["age"].map(age_map)
```

```python
# Verify Conversion
df["age"].unique()
```

```python
# Check Data Type After Conversion
df["age"].dtype
```

## **8. Diagnosis Columns (diag_1, diag_2, diag_3)**

```python
# Check Missing Values
df[["diag_1", "diag_2", "diag_3"]].isnull().sum()
```

```python
# Check Unique Values
df["diag_1"].unique()
df["diag_2"].unique()
df["diag_3"].unique()
```

```python
# Replace '?' with NaN and Fill Missing Values

df["diag_1"] = df["diag_1"].replace("?", np.nan)
df["diag_1"] = df["diag_1"].fillna("Unknown")

df["diag_2"] = df["diag_2"].replace("?", np.nan)
df["diag_2"] = df["diag_2"].fillna("Unknown")

df["diag_3"] = df["diag_3"].replace("?", np.nan)
df["diag_3"] = df["diag_3"].fillna("Unknown")
```

```python
# Verify Missing Values
df[["diag_1", "diag_2", "diag_3"]].isnull().sum()
```

## **9. Binary Columns**

```python
# Check Unique Values
df["change"].unique()
df["diabetesMed"].unique()
```

```python
# Convert Binary Values into Numerical Values
binary_map = {
    "No": 0,
    "Yes": 1,
    "Ch": 1
}

df["change"] = df["change"].map(binary_map)
df["diabetesMed"] = df["diabetesMed"].map(binary_map)
```

```python
# Verify Conversion
df["change"].unique()
df["diabetesMed"].unique()
```

## **10. Target Column**

```python
# Check Data Type
df["readmitted"].dtype
```

```python
# Check Unique Values
df["readmitted"].unique()
```

```python
# Convert Target Values into Numerical Labels
target_map = {
    "NO": 0,
    ">30": 1,
    "<30": 2
}

df["readmitted"] = df["readmitted"].map(target_map)
```
