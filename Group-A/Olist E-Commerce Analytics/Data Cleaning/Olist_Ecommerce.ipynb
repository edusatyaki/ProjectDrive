Merge the table in one big table

```python
products = pd.merge(products, translation, on='product_category_name', how='left')
products
```

```python
products.drop(columns=['product_category_name'], inplace=True)
products.rename(columns={'product_category_name_english': 'product_category'}, inplace=True)
```

```python
df = pd.merge(orders, customers, on='customer_id', how='left')
df = pd.merge(df, items, on='order_id', how='left')
df = pd.merge(df, payments, on='order_id', how='left')
df = pd.merge(df, reviews, on='order_id', how='left')
df = pd.merge(df, sellers, on='seller_id', how='left')
df = pd.merge(df, products, on='product_id', how='left')
```

```python
df
```

```python
df = pd.merge(df, geolocation, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='left')
```

```python
df.drop(columns=['geolocation_zip_code_prefix'], inplace=True)
```

```python
columns_to_drop = [
    'review_id',
    'review_comment_title',
    'review_comment_message',
    'customer_unique_id',
    'customer_zip_code_prefix',
    'seller_zip_code_prefix',
    'order_item_id'
]
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

```python
df
```

```python
extra_columns_to_drop = [
    'product_name_lenght',
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm',
    'geolocation_city',
    'geolocation_state',
    'payment_sequential'
]

df.drop(columns=extra_columns_to_drop, inplace=True, errors='ignore')
```

```python
df['payment_installments'] = df['payment_installments'].fillna(1)
```

```python
df['review_creation_date'] = df['review_creation_date'].fillna('No Review Sent')
df['review_answer_timestamp'] = df['review_answer_timestamp'].fillna('No Answer')
```

```python
df['review_score'] = df['review_score'].fillna(-1)
```

```python
df['payment_installments'] = df['payment_installments'].fillna(1)
```
