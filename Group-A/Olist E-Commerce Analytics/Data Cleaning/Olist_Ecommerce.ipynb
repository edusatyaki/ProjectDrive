## Data Cleaning and Preprocessing

In this section, we will clean and preprocess the loaded datasets. This involves:
1.  **Checking for Missing Values**: Identify and handle null values in each dataset.
2.  **Checking for Duplicates**: Identify and remove duplicate rows.
3.  **Correcting Data Types**: Ensure that columns have appropriate data types (e.g., converting date strings to datetime objects).
4.  **Handling Outliers**: (If necessary) Address any outliers that could skew analysis or modeling.

Let's start with `Customers` dataset.

```python
Customers = Customers.drop_duplicates()
```

```python
Customers['customer_city'] = Customers['customer_city'].str.lower().str.strip()
```

```python
Customers['customer_zip_code_prefix'] = (
    Customers['customer_zip_code_prefix']
    .astype(float)
    .astype(int)
    .astype(str)
    .str.zfill(5)
)
print("Sample ZIPs:", Customers['customer_zip_code_prefix'].head())
```

```python
dataframes = {
    'orders'       : orders,
    'order_items'  : order_items,
    'order_payment': order_payment,
    'order_review' : order_review,
    'products'     : products,
    'sellers'      : sellers,
    'geolocation'  : geolocation,
    'product_category_name': product_category_name
}

for name, df in dataframes.items():
    print(f"\n{'='*50}")
    print(f"  {name.upper()}")
    print(f"{'='*50}")
    print(f"Shape     : {df.shape}")
    print(f"Columns   : {list(df.columns)}")
    print(f"\nNull counts:\n{df.isnull().sum()}")
    print(f"\nDtypes:\n{df.dtypes}")
    print(f"\nPreview:")
    display(df.head(3))
```

```python
# NULL HANDLING

# products — 8 columns have nulls
print("Products nulls:\n", products.isnull().sum())
# Fill missing product name/category with 'unknown'
products['product_category_name']          = products['product_category_name'].fillna('unknown')
products['product_name_lenght']            = products['product_name_lenght'].fillna(0)
products['product_description_lenght']     = products['product_description_lenght'].fillna(0)
products['product_photos_qty']             = products['product_photos_qty'].fillna(0)
products['product_weight_g']               = products['product_weight_g'].fillna(products['product_weight_g'].median())
products['product_length_cm']              = products['product_length_cm'].fillna(products['product_length_cm'].median())
products['product_height_cm']              = products['product_height_cm'].fillna(products['product_height_cm'].median())
products['product_width_cm']               = products['product_width_cm'].fillna(products['product_width_cm'].median())
print("\nProducts nulls after fix:\n", products.isnull().sum())

# orders — 3 timestamp columns have nulls (cancelled orders — expected)
print("\nOrders nulls:\n", orders.isnull().sum())
# These NaNs are valid (cancelled orders have no delivery date) — just document them
print("Note: Null timestamps in orders = cancelled/undelivered orders. Kept as NaN intentionally.")

# order_review — 2 comment columns have nulls
print("\nOrder review nulls:\n", order_review.isnull().sum())
order_review['review_comment_title']   = order_review['review_comment_title'].fillna('no title')
order_review['review_comment_message'] = order_review['review_comment_message'].fillna('no comment')
print("Review comment nulls filled with placeholder text.")
```

```python
orphan_items = ~order_items['order_id'].isin(orders['order_id'])
print("Orphan order_items:", orphan_items.sum())
```

```python
orphan_payments = ~order_payment['order_id'].isin(orders['order_id'])
print("Orphan payments:", orphan_payments.sum())
```

```python
orphan_reviews = ~order_review['order_id'].isin(orders['order_id'])
print("Orphan reviews:", orphan_reviews.sum())
```

```python
orphan_orders = ~order_items['order_id'].isin(orders['order_id'])
print("Orphan orders:", orphan_orders.sum())
```

```python
orphan_sellers = ~order_items['seller_id'].isin(sellers['seller_id'])
print("Orphan sellers:", orphan_sellers.sum())
```

```python
orphan_products = ~order_items['product_id'].isin(products['product_id'])
print("Orphan products:", orphan_products.sum())
```

```python
print("\nAll relationship checks done!")
```

```python
datetime_cols_orders = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
for col in datetime_cols_orders:
    orders[col] = pd.to_datetime(orders[col])

order_review['review_creation_date'] = pd.to_datetime(order_review['review_creation_date'])
order_review['review_answer_timestamp'] = pd.to_datetime(order_review['review_answer_timestamp'])

order_items['shipping_limit_date'] = pd.to_datetime(order_items['shipping_limit_date'])

print("Dtypes after parsing:")
print(orders[datetime_cols_orders].dtypes)
```

```python
payments_agg = order_payment.groupby('order_id').agg(
    total_payment_value = ('payment_value', 'sum'),
    payment_installments = ('payment_installments', 'max'),
    payment_types_used  = ('payment_type', lambda x: ','.join(sorted(x.unique())))
).reset_index()

print("Original payment rows:", len(order_payment))
print("After aggregation (one per order):", len(payments_agg))
print(payments_agg.head())
```

```python
print("order_review rows before dedup:", len(order_review))
order_review = (
    order_review
    .sort_values('review_answer_timestamp')
    .drop_duplicates(subset='order_id', keep='last')
)
print("order_review rows after dedup :", len(order_review))
```

```python
master = orders.merge(Customers, on='customer_id', how='left')
master = master.merge(payments_agg, on='order_id', how='left')
master = master.merge(
    order_items.groupby('order_id').agg(
        total_items       = ('order_item_id', 'count'),
        total_freight     = ('freight_value', 'sum'),
        total_price       = ('price', 'sum')
    ).reset_index(),
    on='order_id', how='left'
)
master = master.merge(
    order_review[['order_id','review_score']],
    on='order_id', how='left'
)

print("Master table shape:", master.shape)
print(master.head())
```

```python
master['delivery_days'] = (
    master['order_delivered_customer_date'] - master['order_purchase_timestamp']
).dt.days


master['delay_days'] = (
    master['order_delivered_customer_date'] - master['order_estimated_delivery_date']
).dt.days

master['is_late'] = master['delay_days'].apply(
    lambda x: True if pd.notna(x) and x > 0 else (False if pd.notna(x) else np.nan)
)

print("Sample engineered features:")
print(master[['order_id','delivery_days','delay_days','is_late']].head(10))
print("\nLate delivery rate:", master['is_late'].mean().round(3))
```

```python
# Category dtype for low-cardinality string columns
master['customer_state']  = master['customer_state'].astype('category')
master['order_status']    = master['order_status'].astype('category')
master['payment_types_used'] = master['payment_types_used'].astype('category')


float_cols = ['total_price', 'total_freight', 'total_payment_value',
              'delivery_days', 'delay_days']
for col in float_cols:
    master[col] = pd.to_numeric(master[col], downcast='float')

print("Dtypes after optimization:")
print(master[['customer_state','order_status','total_price','delivery_days']].dtypes)
```

```python
print("="*50)
print("       MASTER TABLE — FINAL CHECKS")
print("="*50)
print(f"\nShape                  : {master.shape}")
print(f"\nNull counts:\n{master.isnull().sum()}")
print(f"\nDuplicate rows         : {master.duplicated().sum()}")
print(f"\nLate delivery rate     : {master['is_late'].mean():.1%}")
print(f"Avg delivery time      : {master['delivery_days'].mean():.1f} days")

master.to_csv('/content/drive/MyDrive/Ecommerce/master_cleaned.csv', index=False)
print("\nmaster_cleaned.csv saved to Google Drive!")
```
