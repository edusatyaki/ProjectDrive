# **About the Dataset**

The Diabetes 130-US Hospitals Dataset contains over 100,000 hospital admissions of diabetic patients collected from 130 hospitals in the United States between 1999 and 2008.

Each record represents one hospital encounter and contains demographic information, admission details, diagnoses, laboratory procedures, medications, length of stay, and the patient's readmission status.

The primary objective of this project is to clean and prepare the dataset for exploratory analysis and predictive modeling to identify the factors associated with hospital readmission.

## Cleaning Strategy Proposed

Based on the initial data quality assessment and domain understanding, the following data cleaning and preparation steps are proposed for this dataset:

*   **Missing Value Identification**: Identify '?' characters (e.g., in `race`, `weight`, `payer_code`, `medical_specialty`, `diag_1`, `diag_2`, `diag_3`) as missing values and replace them with `np.nan` for consistent handling across the dataset.
*   **Feature Enrichment**: Merge the `diabetic_df` with the `IDS_mapping.csv` file (if necessary, or ensure descriptions are correctly linked/available) to obtain descriptive names for `admission_type_id`, `discharge_disposition_id`, and `admission_source_id`. This enhances readability and interpretability.
*   **High Missing Value Columns**: Consider dropping columns with a very high percentage of missing values (e.g., `weight`, `max_glu_serum`, `A1Cresult`), as they may contain insufficient information to be useful for analysis or imputation. A threshold for dropping will be determined.
*   **Categorical Missing Value Imputation**: For remaining categorical columns with missing values (e.g., `admission_type_description`, `discharge_description`, `admission_source_description` and others after '?' replacement), `np.nan` values will be imputed with 'Unknown' to preserve the records and signify their unknown status. This prevents loss of potentially valuable data rows.
*   **Inconsistent Data Handling**: Address any 'Not Mapped' values (e.g., in `discharge_description`, `admission_source_description`) by treating them similarly to other missing values, replacing them with 'Unknown' or appropriate `np.nan` values.

The IDs to remove are:

- **18**: Likely corresponds to missing/NaN descriptions, representing unclear discharge
- **25**: 'Not Mapped'

### Step 4: Identify and handle 'Not Mapped' or 'Not Available' admission sources

Based on the unique values, we observe descriptions like ' Not Mapped' and ' Not Available'. These represent admission sources that are either undefined or not relevant for direct analysis. We will identify the `admission_source_id` values associated with these descriptions and either remove them or group them into a single 'Unknown' category to streamline the data.

```python
print("Unique admission_source_ids and their corresponding descriptions:")
# Create a mapping from ID to description for easier lookup
id_to_description = df.groupby('admission_source_id')['admission_source_description'].apply(lambda x: x.mode()[0] if not x.mode().empty else None)

for id_val, desc_val in id_to_description.items():
    print(f"ID: {id_val}, Description: {desc_val}")

# Identify IDs that correspond to 'Not Mapped' or 'Not Available' descriptions
# Note: Stripping whitespace from descriptions due to observed leading spaces
ids_to_reclassify = df[df['admission_source_description'].isin([' Not Mapped', ' Not Available'])]['admission_source_id'].unique()

print(f"\nAdmission Source IDs corresponding to 'Not Mapped' or 'Not Available': {ids_to_reclassify}")
```

### Step 5: Consolidate 'Not Mapped', 'Not Available' and 'None' descriptions into an 'Unknown' category

We will replace `admission_source_id` values 9, 17, and 20 with a new ID (e.g., 26) that will represent an 'Unknown' admission source. This ensures consistency and better interpretability for our analysis. After this, we will update the `admission_source_description` column to reflect these changes.

```python
# Replace admission_source_id 9, 17, and 20 with a new ID for 'Unknown'
ids_to_replace = [9, 17, 20]
unknown_id = 26 # Choose an ID not currently in use
df['admission_source_id'] = df['admission_source_id'].replace(ids_to_replace, unknown_id)

# Update the admission_source_description column to reflect 'Unknown' for the new ID
df.loc[df['admission_source_id'] == unknown_id, 'admission_source_description'] = 'Unknown'

print("Value counts for 'admission_source_id' after consolidation:")
df['admission_source_id'].value_counts().sort_index()

print("\nValue counts for 'admission_source_description' after consolidation:")
df['admission_source_description'].value_counts()
```
